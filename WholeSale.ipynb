```python
from sklearn.impute import SimpleImputer
# Define numeric and categorical columns
numeric_columns = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicatessen']
categorical_columns = ['Channel', 'Region']

# Replace non-numeric values (like '?') in numeric columns with NaN
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Impute missing values in numeric columns with mean
imputer_numeric = SimpleImputer(strategy='mean')
df[numeric_columns] = imputer_numeric.fit_transform(df[numeric_columns])

# Impute missing values in categorical columns with most frequent value
imputer_categorical = SimpleImputer(strategy='most_frequent')
df[categorical_columns] = imputer_categorical.fit_transform(df[categorical_columns])

# Verify no more null values
print("Null values after handling:\n", df.isnull().sum())
```

#### Handle Outliers

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range    

for i in df.columns:
    if df[i].dtype !='object':
        lr,ur=remove_outlier(df[i])
        df[i]=np.where(df[i]>ur,ur,df[i])
        df[i]=np.where(df[i]<lr,lr,df[i])
df.head()
```

```python
# Final check after cleaning steps
print("Shape of cleaned dataset:", df.shape)
print("Summary statistics after cleaning:\n", df.describe())
```

# 4) Customer Segmentation

### 1) Can buyers be grouped into segments based on their spending patterns? (e.g. using clustering analysis)


```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer
```

```python
# Assuming df is your DataFrame and it contains spending data
spending_data = df[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicatessen']]

# Impute missing values with the mean of each column
imputer = SimpleImputer(strategy='mean')
imputed_data = imputer.fit_transform(spending_data)
# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(imputed_data)

# Determine the optimal number of clusters using the Elbow method
ssd = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(scaled_data)
    ssd.append(kmeans.inertia_)
# Plot the results
plt.figure(figsize=(8, 4))
plt.plot(K, ssd, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared distances')
plt.title('Elbow Method For Optimal Number of Clusters')
plt.show()
```

```python
# Fit the K-Means model with an appropriate number of clusters
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(scaled_data)
# Add the cluster labels to the original DataFrame
df['Cluster'] = clusters
# Print the mean spending for each cluster
cluster_centers = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=spending_data.columns)
cluster_centers['Cluster'] = range(3)
print(cluster_centers)

# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Grocery', y='Milk', hue='Cluster', data=df, palette='viridis')
plt.title('Clustering of Buyers Based on Spending Patterns')
plt.show()
```

# 7) Buyer Insights

### 1) What is the repeat purchase rate for buyers who spend above the average in at least three categories?

```python
# Calculate the average spending for each category
category_means = df[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicatessen']].mean()

# Identify buyers who spend above the average in at least three categories
def above_average_count(row):
    count = 0
    for category in ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicatessen']:
        if row[category] > category_means[category]:
            count += 1
    return count

# Apply the function to each row
df['Above_Average_Count'] = df.apply(above_average_count, axis=1)

# Identify buyers who spend above the average in at least three categories
repeat_buyers = df[df['Above_Average_Count'] >= 3]

# Calculate the repeat purchase rate
repeat_purchase_rate = len(repeat_buyers) / len(df)

# Print the result
print("Repeat purchase rate for buyers who spend above the average in at least three categories:", repeat_purchase_rate)
```

### 2) How many buyers spend consistently (i.e., similar amounts) across all categories?

```python
# Calculate the standard deviation of spending across all categories for each buyer
df['Spending_StdDev'] = df[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicatessen']].std(axis=1)

# Define a threshold for consistency. Here, we'll assume a standard deviation of less than a certain value indicates consistent spending.
# The threshold can be adjusted based on the specific requirements or distribution of the standard deviations.
threshold = df['Spending_StdDev'].quantile(0.25)  # For example, using the 25th percentile as a threshold

# Identify buyers who spend consistently across all categories
consistent_spenders = df[df['Spending_StdDev'] < threshold]

# Count the number of consistent spenders
num_consistent_spenders = len(consistent_spenders)

print("Number of buyers who spend consistently across all categories:", num_consistent_spenders)
```

### 3)	Which region has the most diverse spending patterns (i.e., high variance in spending across categories)?

```python
# Calculate the variance of spending across all categories for each region
region_variances = df.groupby('Region')[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicatessen']].var()

# Calculate the average variance for each region
region_avg_variances = region_variances.mean(axis=1)

# Identify the region with the highest average variance
most_diverse_region = region_avg_variances.idxmax()

print("The region with the most diverse spending patterns is:", most_diverse_region)
print("Average variances per region:\n",region_avg_variances)
```
